```python
import pandas as pd
import plotly.graph_objects as go


df =pd.read_csv('f_3m.csv')
df2= df.iloc[-850:-300]

candlestic =  go.Candlestick( x = df2['open_time'],open=df2['open'],  high= df2['high'], low =df2['low'],close=df2['close'] ) 
fig = go.Figure( data = [candlestic])

fig.show()
```
